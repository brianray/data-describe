# Data Heatmap

```python
import pandas as pd
from mwdata import data_heatmap
```

```python
from sklearn.datasets import load_wine
data = load_wine()
df = pd.DataFrame(data.data, columns=list(data.feature_names))
df['target'] = data.target
```

## Basic heatmap

```python
data_heatmap(df)
```

## Static heatmap plot

```python
data_heatmap(df, interactive=False)
```

## Missing values only

```python
#Create missing values
df['hue'] = df['hue'].map(lambda x: None if x < 1.05 else x)
df['magnesium'] = df['magnesium'].map(lambda x: None if x%2 ==0 else x)
```

```python
data_heatmap(df, missing=True)
```
