# Feature Importance

```python
import pandas as pd
import mwdata as mw

from sklearn.naive_bayes import GaussianNB
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv('../data/er_data.csv')
df.drop('number_inpatient',axis=1, inplace=True)
df.head(1)
```

## Default
Pass in the data frame and the column name of the response variable

```python
mw.importance(df, 'readmitted')
```

## Show negative importance
Use `truncate=False` to disable removal of negative importance values.

```python
mw.importance(df, 'readmitted', truncate=False)
```

## Getting the data values
You can get the importance values by setting `return_values` to `True`.

```python
mw.importance(df, 'readmitted', return_values=True)
```

## Alternate Model
You can also specify a different model type using the `estimator` argument.

```python
mw.importance(df, 'readmitted', estimator=GaussianNB())
```
