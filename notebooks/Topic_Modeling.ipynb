# Topic Modeling

```python
import pandas as pd
```

```python
%%capture

from mwdata.text import topic_model
```

```python
from sklearn.datasets import fetch_20newsgroups
categories = ['alt.atheism']
newsgroups_train = fetch_20newsgroups(subset='train', categories=categories)
```

```python
df = pd.DataFrame(newsgroups_train['data'])
```

## Explicitly providing number of topics

```python
lda_model = topic_model.TopicModel(num_topics=2)
lda_model.fit(df[0])
```

```python
lda_model.show()
```

## Guess optimal number of topics and show elbow plot

```python
lda_model = topic_model.TopicModel(num_topics=None)
lda_model.fit(df[0])
```

```python
lda_model.show()
```

```python
lda_model.show(display_item='elbow')
```
