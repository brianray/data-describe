# Distributions

```python
import pandas as pd
import mwdata as mw
```

```python
df = pd.read_csv("../data/weatherAUS.csv")
```

```python
df.head(2)
```

*Pick relevant columns here*

```python
df = df[['Location','Rainfall','Cloud9am','WindGustDir', 'RainTomorrow']]
```

## Using Defaults

```python
mw.distribution(df)
```

## Disabling combination of categorical features with many unique values

```python
mw.distribution(df, max_categories=None)
```

## Plotting *all* variables and combinations

```python
mw.distribution(df, plot_all=True, max_categories=5)
```

## Passing keyword arguments to the underlying plotting functions (Seaborn library)

```python
mw.distribution(df, max_categories=3, hist_kwargs={'vertical': True}, violin_kwargs={'hue': 'RainTomorrow'})
```
